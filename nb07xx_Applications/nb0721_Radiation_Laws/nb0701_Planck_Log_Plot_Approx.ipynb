# Planck's Law of Black Body Radiation #

```python
%matplotlib inline
#import matplotlib as mpl
from matplotlib import pyplot as plt

import numpy as np
from numpy import pi, exp
```

```python
plt.rcParams['figure.figsize'] = (4,3) # Set the default figure size. You have to calibrate DPI
plt.rcParams['figure.dpi'] = 136 # depends on your system
#mpl.rcParams['figure.figsize'] = (6,4) # Set the default figure size. You have to calibrate DPI
#mpl.rcParams['figure.dpi'] = 300 # depends on your system
plt.show()
```

```python
def planck(T,lam):
    from scipy.constants import k,h,c
    #s = sigma
    # here we could also import all the constants from scipy.constants
    #k = 1.38054e-23 # J/K
    #h = 6.6256e-34 # Js
    #s = 5.6703e-8 # W/(m^2 K^4)
    #c = 2.997e8 # m/s
    c1= 2*pi*c**2*h
    c2= c*h/k
    el = c1/(lam**5*(exp(c2/(lam*T))-1))
    return(el)
```

```python
lam=np.arange(1,10000)*1e-9 # lam = 0 nm causes problems in the function planck()
```

```python
lam
```

```python
T1 = 6000
e1=planck(T1,lam)

T2 = 3000
e2 = planck(T2,lam)

T3 = 300
e3=planck(T3,lam)
```

### What the students calculated in the Physics 2 class ###

```python
# Results form the Students' trapezoid area calculations:
A1 = 19.97e6 # W/m
A2 = 34.3e6
A3 = 8.4e6
A4 = 8.4e6
A5 = 5.5e6
A = A1 + A2 +A3 + A4 + A5

T = 6000 # K
sigma = 5.67e-8 # W/(m^2 K^4)

eT = sigma * T**4

print ("Approximation (sum of trapezoids):  A = %.3g W/m²" % A)
print ("Stefan-Boltzmann:                  eT = %.3g W/m²" % eT)
```

```python
absolute_error = A-eT # is it big or small? It depends!
relative_error = (A-eT)/eT # deviation in relation to the real value

print("The absolute error is %10.3g W/m²" % absolute_error)
print("The relative error is %7.3f" % relative_error)
print("The relative error is %+5.1f %%" % (relative_error*100))
```
